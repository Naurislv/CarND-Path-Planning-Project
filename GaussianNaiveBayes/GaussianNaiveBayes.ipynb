```python
import pandas as pd
from IPython.display import display
```

```python
train_labels_path = 'nd013_pred_data/train_labels.txt'
train_states_path = 'nd013_pred_data/train_states.txt'

test_labels_path = 'nd013_pred_data/test_labels.txt'
test_states_path = 'nd013_pred_data/test_states.txt'

state_names = ['s', 'd', 's_dot', 'd_dot']
label_names = ['movement']

train_labels = pd.read_csv(train_labels_path, names=label_names)
train_states = pd.read_csv(train_states_path, names=state_names)

test_labels = pd.read_csv(test_labels_path, names=label_names)
test_states = pd.read_csv(test_states_path, names=state_names)

display(train_labels.head(), test_labels.head())
display(train_states.head(), test_states.head())
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
clf.fit(train_states.as_matrix(), train_labels.as_matrix().ravel())

# Test data score
clf.score(test_states.as_matrix(), test_labels.as_matrix().ravel())
```
